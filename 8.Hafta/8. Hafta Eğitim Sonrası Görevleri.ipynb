# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def vki_hesapla(boy, kilo):
    """Vücut Kitle İndeksi hesaplar."""
    return kilo / (boy ** 2)

def vki_sinifi(vki):
    """VKİ'ye göre sınıflandırma yapar."""
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 24.9:
        return "Normal kilolu"
    elif 25 <= vki < 29.9:
        return "Fazla kilolu"
    elif 30 <= vki < 39.9:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veri_kaydet(dosya_adi, bilgiler):
    """Bilgileri dosyaya kaydeder."""
    with open(dosya_adi, "a") as dosya:
        for bilgi in bilgiler:
            dosya.write(bilgi + "\n")

def main():
    giriş_sayısı = int(input("Kaç kişinin verisini girmek istiyorsunuz? "))

    bilgiler = []
    
    for _ in range(giriş_sayısı):
        print("\nKişi bilgilerini girin:")
        isim = input("İsim: ")
        soyisim = input("Soyisim: ")
        yaş = int(input("Yaş: "))
        boy = float(input("Boy (metre cinsinden): "))
        kilo = float(input("Kilo (kg): "))

        vki = vki_hesapla(boy, kilo)
        sınıf = vki_sinifi(vki)
    
        bilgi = (f"İsim: {isim} {soyisim}, Yaş: {yaş}, Boy: {boy}m, Kilo: {kilo}kg, "
                 f"VKİ: {vki:.2f}, Sınıf: {sınıf}")
        bilgiler.append(bilgi)
    
  
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime
menü = {
    "Mango Smoothie": 5.00,
    "Berry Blast": 4.50,
    "Green Detox": 6.00,
    "Tropical Punch": 5.50,
    "Protein Power": 7.00
}

satilan_icecekler = []

def menuyu_goruntule():
    """Menüyü görüntüler."""
    print("\nMenü:")
    for icecek, fiyat in menü.items():
        print(f"{icecek}: ${fiyat:.2f}")

def siparis_al():
    """Kullanıcıdan içecek siparişi alır ve fiyatı hesaplar."""
    toplam_fiyat = 0
    siparis = []
    
    while True:
        menuyu_goruntule()
        secim = input("\nSipariş vermek istediğiniz içeceği girin (Çıkmak için 'q'): ").strip()
        
        if secim.lower() == 'q':
            break
        
        if secim in menü:
            siparis.append(secim)
            toplam_fiyat += menü[secim]
            print(f"{secim} eklendi. Toplam fiyat: ${toplam_fiyat:.2f}")
        else:
            print("Geçersiz içecek adı. Lütfen menüdeki içeceklerden birini seçin.")
    
    if siparis:
        tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        kayit = {
            "tarih": tarih,
            "icerikler": siparis,
            "toplam_fiyat": toplam_fiyat
        }
        satilan_icecekler.append(kayit)
        print(f"\nSiparişiniz alındı. Toplam fiyat: ${toplam_fiyat:.2f}")
        print("Sipariş bilgileri kaydedildi.")
    else:
        print("Sipariş verilmedi.")

def siparisleri_goruntule():
    """Satın alınan içeceklerin kayıtlarını görüntüler."""
    if not satilan_icecekler:
        print("\nHenüz hiçbir sipariş verilmedi.")
    else:
        print("\nSatın Alınan İçcekler:")
        for siparis in satilan_icecekler:
            print(f"Tarih: {siparis['tarih']}")
            print("İçecekler:")
            for icerik in siparis['icerikler']:
                print(f"  - {icerik}")
            print(f"Toplam Fiyat: ${siparis['toplam_fiyat']:.2f}")
            print()

def main():
    """Ana fonksiyon: Menü ve sipariş işlemlerini yönetir."""
    while True:
        print("\nYetBar - Smoothie Bank")
        print("1. Menüyü Görüntüle")
        print("2. Sipariş Ver")
        print("3. Satın Alınan İçcekleri Görüntüle")
        print("4. Çıkış")
        
        secim = input("Bir seçenek girin (1/2/3/4): ").strip()
        
        if secim == '1':
            menuyu_goruntule()
        elif secim == '2':
            siparis_al()
        elif secim == '3':
            siparisleri_goruntule()
        elif secim == '4':
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçenek. Lütfen 1, 2, 3 veya 4 girin.")

if __name__ == "__main__":
    main()
```
