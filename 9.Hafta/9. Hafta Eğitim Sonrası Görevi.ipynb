# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self, tv_durum="Kapalı", tv_ses=0, kanal_listesi=None, aktif_kanal=None):
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi if kanal_listesi is not None else []
        self.aktif_kanal = aktif_kanal

    def tv_ac(self):
        if self.tv_durum == "Açık":
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            time.sleep(1)
            self.tv_durum = "Açık"
            print("Televizyon açık.")

    def tv_kapat(self):
        if self.tv_durum == "Kapalı":
            print("Televizyon zaten kapalı.")
        else:
            print("Televizyon kapanıyor...")
            time.sleep(1)
            self.tv_durum = "Kapalı"
            print("Televizyon kapalı.")

    def ses_ayarlari(self):
        while True:
            secim = input("Sesi Arttır: '+'\nSesi Azalt: '-'\nÇıkış: 'q'\nSeçiminiz: ")
            if secim == "+":
                if self.tv_ses < 100:
                    self.tv_ses += 1
                    print("Ses:", self.tv_ses)
                else:
                    print("Ses en yüksek seviyede.")
            elif secim == "-":
                if self.tv_ses > 0:
                    self.tv_ses -= 1
                    print("Ses:", self.tv_ses)
                else:
                    print("Ses en düşük seviyede.")
            elif secim == "q":
                print("Ses ayarlarından çıkılıyor.")
                break
            else:
                print("Geçersiz bir seçim yaptınız.")

    def kanal_ekle(self, kanal_ismi):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal_ismi)
        print(f"{kanal_ismi} kanalı eklendi.")

    def kanal_sayisini_ogrenme(self):
        return len(self.kanal_listesi)

    def rastgele_kanal_gecme(self):
        if self.kanal_listesi:
            self.aktif_kanal = random.choice(self.kanal_listesi)
            print(f"Şu anki kanal: {self.aktif_kanal}")
        else:
            print("Henüz kanal eklenmemiş.")

    def televizyon_bilgileri(self):
        print(f"Televizyon Durumu: {self.tv_durum}\nSes: {self.tv_ses}\nKanal Sayısı: {self.kanal_sayisini_ogrenme()}\nAktif Kanal: {self.aktif_kanal}")

def main():
    kumanda = Kumanda()

    print("""
    *** TELEVİZYON UYGULAMASI ***

    1. Tv Aç
    2. Tv Kapat
    3. Ses Ayarları
    4. Kanal Ekle
    5. Kanal Sayısını Öğrenme
    6. Rastgele Kanala Geçme
    7. Televizyon Bilgileri
    Çıkış için 'q' tuşuna basın.
    """)

    while True:
        islem = input("İşlemi Seçiniz: ")

        if islem == "q":
            print("Programdan çıkılıyor...")
            break
        elif islem == "1":
            kumanda.tv_ac()
        elif islem == "2":
            kumanda.tv_kapat()
        elif islem == "3":
            kumanda.ses_ayarlari()
        elif islem == "4":
            kanal_ismi = input("Eklenecek Kanal İsmi: ")
            kumanda.kanal_ekle(kanal_ismi)
        elif islem == "5":
            print("Kanal Sayısı:", kumanda.kanal_sayisini_ogrenme())
        elif islem == "6":
            kumanda.rastgele_kanal_gecme()
        elif islem == "7":
            kumanda.televizyon_bilgileri()
        else:
            print("Geçersiz işlem!")

if __name__ == "__main__":
    main()

```
